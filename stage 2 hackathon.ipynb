```python
import json

# Assuming the data is already loaded into a variable named 'data'

all_key_value_pairs = {}

def extract_key_value_pairs(data):
  """
  This function recursively extracts key-value pairs from a dictionary.

  Args:
    data: The data to extract key-value pairs from (can be a dictionary or a list).

  Returns:
    A dictionary containing all extracted key-value pairs.
  """
  if isinstance(data, dict):
    for key, value in data.items():
      all_key_value_pairs[key] = value
      extract_key_value_pairs(value)  # Recursively extract from nested dictionaries
  elif isinstance(data, list):
    for i, item in enumerate(data):
      all_key_value_pairs[f"item_{i}"] = item  # Use item index for list elements

# Call the function to extract key-value pairs
extract_key_value_pairs(data["design_state_data"]["algorithms"])

# Now all_key_value_pairs contains all key-value pairs from the "algorithms" section

print(all_key_value_pairs)
```

# Extracting all the necessary models and saving them in a text file

```python
def get_selected_model_text(data):
  """
  This function creates a simple text representation of selected models.

  Args:
    data: The dictionary containing information for selected models.

  Returns:
    A string containing text representation of selected models.
  """
  text = ""
  for model_name, model_info in data.items():
    text += f"Model Name: {model_name}\n"
    for key, value in model_info.items():
      if key != "is_selected":  # Skip "is_selected" key
        text += f"\t{key}: {value}\n"
    text += "\n"  # Add newline between models
  return text

# Get selected models info (assuming you have the code from previous step)
selected_models_info = get_selected_model_info(data["design_state_data"]["algorithms"])

# Convert to text representation
text_data = get_selected_model_text(selected_models_info)

print(text_data)

# Save the text data (e.g., write to a file)
with open("selected_models.txt", "w") as file:
  file.write(text_data)

print("Selected models information saved to selected_models.txt")
```

# Extracting all the features 

```python
def get_feature_handling_info(data):
  """
  This function extracts all key-value pairs from the "feature_handling" section.

  Args:
    data: The dictionary containing the design state data.

  Returns:
    A dictionary containing all key-value pairs from "feature_handling".
  """
  feature_handling_info = {}
  if "design_state_data" in data and "feature_handling" in data["design_state_data"]:
    for feature_name, feature_details in data["design_state_data"]["feature_handling"].items():
      # Include feature name as a key-value pair
      feature_handling_info[feature_name] = feature_details
      # Add details within each feature
      feature_handling_info.update(feature_details)
  return feature_handling_info

# Assuming you have the data loaded in a variable named 'data'
feature_handling_data = get_feature_handling_info(data)

print(feature_handling_data)
```

```python
def get_selected_feature_info(data):
  """
  This function extracts information for features with "is_selected": True.

  Args:
    data: The dictionary containing feature handling information.

  Returns:
    A dictionary containing information for features with "is_selected": True.
  """
  selected_features = {}
  for key, value in data.items():
    # Check if key is a feature name (not details within a feature)
    if isinstance(value, dict) and "is_selected" in value and value["is_selected"] == True:
      selected_features[key] = value
  return selected_features

# Assuming you have feature_handling_data from previous step
selected_features_info = get_selected_feature_info(feature_handling_data)

print(selected_features_info)
```
